### üèÅ SOLUTION TOFINAL PRACTICE TASK
To perform the requested actions, use the following sequence of Git commands in your terminal, followed by steps to complete the pull request on GitHub's web interface.

**Git Commands**<br>
1. **Create and switch to a new branch**:

   git checkout -b feature-profile

This command creates a new local branch named feature-profile and immediately switches your working directory to it.

2. **Add a profile.md file and write a short bio**:

   Use a text editor or a command-line editor (like nano or vim) to create and add content to the file.

   nano profile.md

   Inside the editor, write your short bio (e.g., "John Doe is a software developer with 5 years of experience. He specializes in front-end technologies and is passionate about open source.") and save the file.

3. **Stage the new file:**

   git add profile.md

This command adds the new profile.md file to the Git staging area. You can use git status to verify it's staged (it will appear in green).

4. **Commit the changes:**

   git commit -m "Add profile.md with short bio"
   
This command records the staged snapshot in the local repository history with a descriptive message.

5. **Push the branch to the remote repository:**

   git push origin feature-profile

The first time you push a new branch, you might need to use the -u flag to set the upstream branch: git push -u origin feature-profile. 

### Open and Merge Pull Request
   The pull request must be opened and merged using the remote repository's web interface (e.g., GitHub, GitLab, Bitbucket), as the command line doesn't natively support all the UI-based PR features like adding reviewers or descriptions. 
1. **Navigate to your repository on GitHub** (or your chosen platform). A yellow banner or a prompt will likely appear, indicating that the feature-profile branch has recent pushes and offering to "Compare & pull request".

2. **Click on the prompt** to open the "Open a pull request" page.

3. **Verify the base branch** (where you want to merge into, likely main or master) and the compare branch (feature-profile).

4. **Provide a title and description** for your pull request.

5. **Click "Create Pull Request"**.

6. On the pull request page, if there are no conflicts, you will see a "**Merge pull request**" (or similar) button.

7. **Click the merge button** and confirm the action to merge your changes into the base branch. You can then typically **delete** the *feature-profile* **branch** from the UI **if it's no longer needed**.
